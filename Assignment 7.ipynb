```python

%%sql
-- 1. (Query #1)
select forename, surname, drivers.driverid, wins.count as "wins"
from drivers
inner join
(select driverid, count(driverid) from results 
 where position = 1 
 group by driverid) as wins
on wins.driverid = drivers.driverid
order by wins.count desc
limit 10
```

```python
%%sql
-- 2. (Query #2)
select drivers.driverref, constructors.name, results.count from drivers
left join
(select driverid, constructorid, count(*) as "count"
 from results
group by constructorid, driverid) as results
on results.driverid = drivers.driverid
left join constructors
on constructors.constructorid = results.constructorid
order by count desc
limit 10
```

```python
%%sql
-- 4. (Query #3)
select drivers.driverref, constructors.name, results.count from drivers
left join
(select driverid, constructorid, count(*) as "count"
 from results
 where statusid = 1
group by constructorid, driverid) as results
on results.driverid = drivers.driverid
left join constructors
on constructors.constructorid = results.constructorid
where count > 0
order by count desc
limit 10
```

```python
%%sql
-- 5. (Query #4)
select driverid, raceid, SUM(milliseconds) as "total milliseconds" from pitstops 
group by driverid, raceid
order by SUM(milliseconds) desc
limit 10
```

```python
%%sql
-- 8. (query #5)

select drivers.driverref, constructors.name, results.count, pits.ms_spent from drivers
left join
(select driverid, constructorid, count(*) as "count"
 from results
 where statusid = 1
group by constructorid, driverid) as results
on results.driverid = drivers.driverid
left join constructors
on constructors.constructorid = results.constructorid
left join 
(select driverid, sum(milliseconds) as "ms_spent" from pitstops
group by driverid) as pits
on pits.driverid = drivers.driverid

where results.count > 0 and ms_spent > 0

order by ms_spent desc
limit 10
```

```python
%%sql
-- 9. (Query #6)

select drivers.driverref, constructors.name, results.count, 
pits.ms_spent, pits_average.avg

from drivers

left join
(select driverid, constructorid, count(*) as "count"
 from results
 where statusid = 1
group by constructorid, driverid) as results
on results.driverid = drivers.driverid

left join constructors
on constructors.constructorid = results.constructorid

left join 
(select driverid, sum(milliseconds) as "ms_spent" from pitstops
group by driverid) as pits
on pits.driverid = drivers.driverid

left join
(select driverid, AVG(sum) as "avg"
from
(select driverid, raceid, SUM(milliseconds) as "sum" from pitstops
group by driverid, raceid) as pit_sum
group by driverid) as pits_average
on drivers.driverid = pits_average.driverid

where 
results.count > 0 and 
ms_spent > 0 and 
pits_average.avg > 0

order by pits_average.avg asc
limit 10



```
